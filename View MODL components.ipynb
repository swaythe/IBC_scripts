```python
import os
import glob

from nilearn import image
from nilearn import plotting
```

```python
# Set the task and algorithm variables and some folders
task = 'raiders'

modl_folder = os.path.join('..', task, '3mm/modl/no_atlas')
```

```python
# Which set of decompositions do you want?
n_comp = 20

decomp_folder = os.path.join(modl_folder, 'ncomp_' + str(n_comp), 'components')
comp_files = sorted(glob.glob(decomp_folder + '/component-*.nii.gz'))
```

```python
comp_files
```

```python
for i, cf in enumerate(comp_files):
    plotting.plot_stat_map(cf, cut_coords=7, display_mode='z', draw_cross=False)
```
